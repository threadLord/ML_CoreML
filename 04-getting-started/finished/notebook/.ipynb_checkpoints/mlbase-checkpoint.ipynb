```python
data = pd.read_json('corpus.json', orient='records')
authors = data.author
freq = authors.value_counts()
```

```python
data.sort_values(by='title')
```

```python
plt.hist(freq, bins=10, normed=1)
plt.show()
```

```python
(n, bins) = np.histogram(freq, bins=10, normed=True)
plt.plot(.5*(bins[1:]+bins[:-1]), n)
plt.show()
```
