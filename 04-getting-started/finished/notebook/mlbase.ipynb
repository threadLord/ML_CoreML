Now, look at some Python syntax.

`if-else` & `None`:

A major syntax difference is the importance of indentation, which replaces `{}` to define closures (blocks).

Also, Python has a builtin `NoneType` to represent no-value. You should use `is` or `is not` to test for it, not `==`, because classes can override `==`.

```python
if authors is None:
    print('authors is None')
else:
    print('authors is not None')
```

Define and call a function: 

Notice the auto-indent on the second line, and you have to unindent to type the third line. Coding convention says to leave a blank line after the function definition, but it's not a syntax rule. You'd probably feel more comfortable with a blank line there, anyway.

```python
def mysum(x, y):
    return x + y
print(mysum(1, 3))
```

For-loop and arrays: 
- Arrays are pretty similar to Swift. You test whether an array is empty just using its name.
- For loops are also similar to Swift, but with the `:` plus indentation syntax.

To make a point about indentation, indent the last statement to match the `print` statement in the loop:

```python
mylist = [1, 2, 3]
mylist.append(4)
if mylist:
    print('mylist is not empty')

for value in mylist:
    print(value)
    
    print('List Length: %d' % len(mylist))
```
